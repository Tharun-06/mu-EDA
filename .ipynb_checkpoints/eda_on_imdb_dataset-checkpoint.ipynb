```python
data['Released_Year']=data['Released_Year'].replace({'PG':1975})
```

```python
data['Released_Year']=data['Released_Year'].astype(int)
```

```python
data.Runtime.nunique()
data.Runtime.unique()
```

```python
data.select_dtypes(include='object').isna().sum()
```

```python
data.select_dtypes(exclude='object').isna().sum()
```

```python
data['Gross']=data['Gross'].str.replace(",","")
data['Gross']=data['Gross'].astype(int)
```

```python
data['Genre_Cleaned']=data['Genre'].fillna('').apply(lambda x: [i.strip() for i in x.split(',')])
genre_dummies=data['Genre_Cleaned'].explode().str.get_dummies().groupby(level=0).sum()
data=pd.concat([data, genre_dummies],axis=1)
```

```python
data['Runtime']=data['Runtime'].str.extract(r'(\d+)\s*min', expand=False)
data['Runtime']=pd.to_numeric(data['Runtime'], errors='coerce')
```

```python
data.drop(columns=['ID','Poster_Link','Overview','Genre'],axis=1,inplace=True)
```

```python
cols=['Meta_score','IMDB_Rating','Runtime','Released_Year','No_of_Votes','Gross']
num_cols=3
num_rows=(len(cols) + num_cols - 1) // num_cols
fig,axes=plt.subplots(num_rows, num_cols, figsize=(6 * num_cols, 4 * num_rows))
axes=axes.flatten()
for i, col in enumerate(cols):
    axes[i].hist(data[col].dropna(), bins=10, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}', fontsize=12)
    axes[i].set_xlabel(col, fontsize=10)
    axes[i].set_ylabel('Frequency', fontsize=10)
    axes[i].tick_params(axis='x', rotation=20)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')
fig.suptitle('Histograms of Selected Movie Features', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
numeric_cols = ['Meta_score','IMDB_Rating','Runtime','Released_Year','No_of_Votes','Gross']
rows = (len(numeric_cols) + 2) // 3
fig, axes = plt.subplots(rows, 3, figsize=(18, 4 * rows))
axes = axes.flatten()
for i, col in enumerate(numeric_cols):
    axes[i].boxplot(data[col].dropna())
    axes[i].set_title(f'Box Plot of {col}')
    axes[i].set_xlabel(col)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')
plt.tight_layout()
plt.show()
numeric_cols
```

```python
# Group and get top 15 directors by average IMDb rating
top15=data.groupby("Director")["IMDB_Rating"].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
colors=plt.cm.viridis(top15.rank(method='first').values/len(top15))  # Normalized rank for gradient
top15.plot(kind="bar", color=colors, edgecolor="black")
plt.ylabel("Average IMDb Rating")
plt.title("Top 15 Directors by Average IMDb Rating")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```
